```python
import requests, json
```

### Get a basic list of all files, and subdirectories, in the study folder

```python
study_id = 'MTBLS1987'
r = requests.get('https://www.ebi.ac.uk:443/metabolights/ws/studies/' + 'MTBLS1987' + '/files/tree?include_sub_dir=true')
r.status_code
```

```python
rjs=r.json()
```

```python
rjs
```

```python
rjs['study']
```

```python
for path_rec in rjs['study']:
    print(path_rec['file'])
```
