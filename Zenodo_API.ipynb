```python
import requests, json
```

```python
r = requests.get('https://zenodo.org/api/records/')
r.status_code
```

```python
r.headers
```

```python
r_body=json.loads(r.text)
print(r_body.keys())
for key, val in r_body.items():
    separ='-'*len(key)
    print(f'{separ}\n{key}\n{separ}')
    print(val)
```

```python
r = requests.get('https://zenodo.org/api/records/4630543')
r.status_code
```

```python
r_body=json.loads(r.text)
print(r_body.keys())
for key, val in r_body.items():
    separ='-'*len(key)
    print(f'{separ}\n{key}\n{separ}')
    print(val)
```

```python
r_body['files']
```

```python
buckets=[]
for file_rec in r_body['files']:
    buckets.append(file_rec['key'])
buckets_unique=set(buckets)
for buck in buckets_unique:
    print(f'{buck} - {buckets.count(buck)}')
```
