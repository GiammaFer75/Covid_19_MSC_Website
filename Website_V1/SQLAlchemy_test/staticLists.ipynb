```python
from openpyxl import Workbook
import re

input_path   = 'C:/COVID_19_MassSpectrometryCoalition/Data/import_dropDown_lst_RAW.txt'
saving_path  = 'C:/COVID_19_MassSpectrometryCoalition/Data/export_dropdown_options.xlsx'

def save_options(output_path, input_path = None, lst = [], regex_pattern='', length=0, length_cond = ''):

    
    def check_writing_conditions():
        
        if (length == 0) and (length_cond == ''): write_line = True
        else: 
            write_line = False
            if (length_cond == '<=') and (len(item) <= length): write_line = True
            if (length_cond == '>=') and (len(item) >= length): write_line = True
            if (length_cond == '==') and (len(item) == length): write_line = True
        #print(write_line)
        return write_line

    
    if regex_pattern:
        extract_pattern = re.compile(r'' + regex_pattern + '') # Compile the pattern for the desired text extraction

    # ***** Read options from the file ***** #
    if input_path:
        with open(input_path) as txt_file:
            [lst.append(line) for line in txt_file.readlines()]                
    
    
    workbook = Workbook()
    sheet = workbook.active
    row_ind = 1
    #print(lst)
    for item in lst:
        # Clean the item if a regex pattern is provided
        if regex_pattern:
            item_extracted = extract_pattern.search(item)
            #item_extracted = extract_pattern.findall(item)
            item = item_extracted.group(1)
        
        # ***** Save the list in xls file ONLY if they pass the control ***** # 
        if check_writing_conditions():     
            cell = sheet.cell(row = row_ind, column = 1)
            print(item)
            cell.value = item
            row_ind += 1
        else:
            print('condition failed')
    workbook.save(filename=output_path)


                        # /////////////////////// ETNICITY \\\\\\\\\\\\\\\\\\\\\\\ #    
# 2021 Census from GOV.UK
# https://www.ethnicity-facts-figures.service.gov.uk/style-guide/ethnic-groups
# I prefer the NHS classification just for the clinical point of view instead of the census point of view

# Etnicity from NHS Digital
# https://digital.nhs.uk/data-and-information/data-collections-and-data-sets/data-sets/mental-health-services-data-set/submit-data/data-quality-of-protected-characteristics-and-other-vulnerable-groups/ethnicity


save_options(output_path=saving_path , input_path=input_path, regex_pattern='', length=3, length_cond = '>=')
```
