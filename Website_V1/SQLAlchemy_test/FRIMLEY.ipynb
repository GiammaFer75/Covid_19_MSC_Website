```python
import pandas as pd

frimley_path = 'C:/COVID_19_MassSpectrometryCoalition/Data/REDCap_Manchester/COVIDMSCoalitionMult_DATA_LABELS_2022-11-02_0942.csv'
frimley_path = 'C:/COVID_19_MassSpectrometryCoalition/Data/REDCap_Manchester/COVIDMSCoalitionMult_DATA_LABELS_2022-12-05_1139.csv'
```

```python
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 30)
#pd.set_option('display.width', 1000) https://stackoverflow.com/questions/11707586/how-do-i-expand-the-output-display-to-see-more-columns-of-a-pandas-dataframe

frimley = pd.read_csv(frimley_path)
frimley
```

```python
#### print(f'Number of columns {len(frimley.columns)} \n')
for ind, col in enumerate(frimley.columns):
    print(f'{ind} - {col}')
```

```python
def columns_group(column_vector, print_dict=True):
    columns_grouped = {}
    group_ind=1
    columns_grouped[group_ind]=[]
    for col_ind, col_name in enumerate(frimley.columns):
        if 'Complete' in col_name:
            columns_grouped[group_ind].append(col_name)
            group_ind += 1
            columns_grouped[group_ind]=[]
        else:
            columns_grouped[group_ind].append(col_name)
    if print_dict:
        for k, v in columns_grouped.items():
            print(f' {k}-\n')
            for value in v: print(f'\t {value}')
    return columns_grouped
# ----------------------------------------------------------    
columns_grouped = columns_group(frimley.columns)
```

```python
from openpyxl import Workbook

saving_path = 'C:/COVID_19_MassSpectrometryCoalition/Data/REDCap_Manchester/columns_grouped.xlsx'

workbook = Workbook()
sheet = workbook.active
s_row, s_col = (0,1)

for group_key, group_values in columns_grouped.items():
    s_row += 1
    cell = sheet.cell(row = s_row, column = s_col)
    cell.value = "Group" + str(group_key)
    s_col += 1
    for column in group_values:
        s_row += 1
        cell = sheet.cell(row = s_row, column = s_col)
        cell.value = column
    s_col -= 1
workbook.save(filename=saving_path)
```
