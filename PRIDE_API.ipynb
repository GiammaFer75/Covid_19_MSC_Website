```python
import requests
# Read this for the returning object details
# https://pypi.org/project/requests/

import sys, json
```

```python
def get_response_header(header_object, exception_nf=False, explore_exception=False):
    for header_key, header_val in header_object.items():# resp.headers
        print(f'{header_key} ----> {header_val}')
        if header_key == 'Access-Control-Allow-Headers':
            allowed_headers_lst = header_val.split(',')
            #print(f'\n {allowed_headers_lst}')
            print()
            for head in allowed_headers_lst:
                try:
                    print(f'{head} === {header_object[head]}') # resp.headers
                except Exception: #### https://wiki.python.org/moin/HandlingExceptions ####
                    # This is how you get the type
                    excType = Exception.__class__.__name__
                    if exception_nf:
                        print(f'\n-------\n{head} -- not found\n-------')
                        e = sys.exc_info()[0]
                        print(f'Exception ===> {e}')
                        if explore_exception:
                            print('Exception object content')
                            for k,v in Exception.__dict__.items():
                                print(f'{k} ---> {v}')
```
